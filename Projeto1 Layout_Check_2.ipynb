```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
#Se usar o arquivo descompactado
data15 = pd.read_csv('PES2015.csv',sep=',')
data11 = pd.read_csv('PES2011.csv',sep=',')
#Se usar o arquivo compactado como está no dropbox púbico
#pd.read_csv('PES2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
#data15
data15 = data15.loc[:,["V4728", "V0701","V3033","V0702","V0703","V0704","V0705"]]
data15 = data15[data15.V4728 >=4]
# data15 = data15[data15.V3033 >=2002]








data15.V0701 = data15.V0701.astype("category")
data15.V0701.cat.categories = ('Sim', 'Não')

data15.V0702 = data15.V0702.astype("category")
data15.V0702.cat.categories = ('Sim','Não')

data15.V0703 = data15.V0703.astype("category")
data15.V0703.cat.categories = ('Sim')

data15.V0704 = data15.V0704.astype("category")
data15.V0704.cat.categories = ('Sim')


# data15 = data15.rename(columns = {"V0701":"Trabalhou no ano", 
# "V9001":"Trabalhou na Semana", "V4803":"Anos de Estudos", "V0601":
# "Sabe ler e escrever"})

data15







```

```python
grafico = data15.V4728.value_counts(True)*100

plt.subplot(121)
plot = grafico.plot(kind ='bar', color = ('red','blue','green','orange'))
plt.title("Distribuição Rural do trabalho infantil")
plt.ylabel("%")

fig = plt.figure(figsize=(12,5))


```

```python
data15agro = data15[data15.V4728 == 8]

data15agrografico = pd.crosstab(data15agro.V0701,data15agro.V0702)

data15agrografico.plot(kind = 'bar')
plt.xlabel('Trabalho Infantil')
plt.title('Distribuição das variáveis na área rural mais povoada')

```
